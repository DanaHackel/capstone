```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
drug_list = []
with open('data/Products.txt', newline = '') as products:                                                                                          
    drug_reader = csv.DictReader(products, delimiter='\t')
    for drug in drug_reader:
        drug_list.append(dict(drug))
drug_df = pd.DataFrame(drug_list)
drug_df
```

```python
drug_df.dtypes
```

```python
drug_df['ApplNo'] = drug_df['ApplNo'].astype(float)
```

```python
drug_df['ProductNo'] = drug_df['ProductNo'].astype(float)
```

```python
drug_df.dropna(axis=1, inplace=True)
```
