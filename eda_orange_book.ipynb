```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import csv
```

```python
drug_list = []
with open('data/products_orange.txt', newline = '') as products:                                                                                          
    drug_reader = csv.DictReader(products, delimiter='~')
    for drug in drug_reader:
        drug_list.append(dict(drug))
orange_df = pd.DataFrame(drug_list)
orange_df
```

```python
#I specifically want the route separate to combine a few columns for one to match an nadac column
#I'll also make a dosage_form one just in case
orange_df['dosage_form'] = orange_df['DF;Route'].str.split(';').str[0]
orange_df['route'] = orange_df['DF;Route'].str.split(';').str[-1]
```

```python
orange_df.head()
```

```python
orange_df['ndc_description_col'] = orange_df['Trade_Name'] + ' ' + orange_df['Strength'] + ' ' + orange_df['route']
orange_df.head()
```
