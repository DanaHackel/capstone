```python
import pandas as pd
import numpy as np
import fuzzywuzzy as fuzzy
```

```python
nadac = pd.read_csv('data/nadac_no_repeats.csv')
orange = pd.read_csv('data/drug_names_df.csv')
```

```python
nadac_names = list(nadac['NDC Description'].unique())
print(f'Unique number of NDC Descriptions in the NADAC dataframe: {len(nadac_names)}')
```

```python
orange_names = list(orange['ndc_description_col'].unique())
print(f'Unique number of Drug Names in the Orange Book dataframe: {len(orange_names)}')
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
from collections import defaultdict
import time

ndc_matches = {}

# tic = time.perf_counter()
# for nadac_name in nadac_names:
#     for orange_name in orange_names:
#         toc = time.perf_counter()
#         score = fuzz.token_sort_ratio(nadac_name, orange_name)
#         if score > 85:
#             ndc_matches[nadac_name] = orange_name
#             toc = time.perf_counter()
#             print(f"{nadac_name} matched to {orange_name} in {toc-tic:0.4f} seconds")
#             continue #so it does not overwrite new pairs, #since I am now working with names instead od numbers, it shouldn't re-write my matches
#         elif score < 85:
#             pass
#         ndc_matches.get(nadac_name, 'No match found')
    
tic = time.perf_counter()
for orange_name in orange_names:
    prev_high_score = 85
    for nadac_name in nadac_names:
        toc = time.perf_counter()
        score = fuzz.token_sort_ratio(nadac_name, orange_name)
        if score > prev_high_score:
            prev_high_score = score
            ndc_matches[nadac_name] = orange_name
            toc = time.perf_counter()
            print(f"{nadac_name} matched to {orange_name} in {toc-tic:0.4f} seconds")
            #so it does not overwrite new pairs, #since I am now working with names instead od numbers, it shouldn't re-write my matches
        else:
            continue
        ndc_matches.get(nadac_name, 'No match found')   
#check for slight mis-matches in both dataframes
```

```python
names_df = pd.DataFrame.from_dict(ndc_matches, orient = 'index', columns=['names'])
names_df.head()
names_df.to_csv('./data/names_df_no_overwrites_long.csv')
```
